```python
import pandas as pd
```

```python
file_path = "Resources/purchase_data.csv"
purchase_data_df = pd.read_csv(file_path)
#purchase_data_df.head()
```

```python
#check date type
#purchase_data_df.dtypes
```

```python
#check na value
#purchase_data_df.count()
```

```python
total_player = len(purchase_data_df["SN"].unique())
total_player_df = pd.DataFrame({"Total Player": [total_player]})
total_player_df_display = total_player_df.copy()
total_player_df_display["Total Player"] = total_player_df_display["Total Player"].map("{:,}".format)
total_player_df_display
```

```python
total_item_count = len(purchase_data_df["Item ID"].unique())
avg_purchase_price = purchase_data_df["Price"].mean()
total_purchase = len(purchase_data_df["Purchase ID"])
total_revenue = purchase_data_df["Price"].sum()
purchase_analysis_total = pd.DataFrame({"Number of Unique Items": [total_item_count],
                                "Average Price": [avg_purchase_price],
                                "Number of Purchases": [total_purchase],
                                "Total Revenue": [total_revenue],
                               })
purchase_analysis_total = purchase_analysis_total.copy()
purchase_analysis_total["Average Price"] = purchase_analysis_total["Average Price"].map("${:,.2f}".format)
purchase_analysis_total["Total Revenue"] = purchase_analysis_total["Total Revenue"].map("${:,.2f}".format)
purchase_analysis_total["Number of Unique Items"] = purchase_analysis_total["Number of Unique Items"].map("{:,}".format)
purchase_analysis_total["Number of Purchases"] = purchase_analysis_total["Number of Purchases"].map("{:,}".format)
purchase_analysis_total
```

```python
gender_data = purchase_data_df[["SN","Gender"]].drop_duplicates()
player_count_gender = gender_data["Gender"].value_counts()
gender_demo_df = pd.DataFrame({"Total Count": player_count_gender})
gender_demo_df["Percentage of Players"] = gender_demo_df["Total Count"]/gender_demo_df["Total Count"].sum()
gender_demo_df_display = gender_demo_df.copy()
gender_demo_df_display["Percentage of Players"] = gender_demo_df_display["Percentage of Players"].map("{:,.2%}".format)
gender_demo_df_display["Total Count"] = gender_demo_df_display["Total Count"].map("{:,}".format)
gender_demo_df_display
```

```python
grouped_gender = purchase_data_df.groupby(["Gender"])
purchase_count_gender = grouped_gender["Purchase ID"].count()
total_revenue_gender = grouped_gender["Price"].sum()
avg_purchase_price_gender = grouped_gender["Price"].mean()
avg_purchase_per_person_gender = (total_revenue_gender / player_count_gender)
purchase_analysis_gender = pd.DataFrame({"Purchase Count": purchase_count_gender,
                                        "Average Purchase Price": avg_purchase_price_gender,
                                        "Total Purchase Value": total_revenue_gender,
                                        "Avg Total Purchase per Person": avg_purchase_per_person_gender,
                                        })
purchase_analysis_gender_display = purchase_analysis_gender.copy()
purchase_analysis_gender_display["Average Purchase Price"] = purchase_analysis_gender_display["Average Purchase Price"].map("${:,.2f}".format)
purchase_analysis_gender_display["Total Purchase Value"] = purchase_analysis_gender_display["Total Purchase Value"].map("${:,.2f}".format)
purchase_analysis_gender_display["Avg Total Purchase per Person"] = purchase_analysis_gender_display["Avg Total Purchase per Person"].map("${:,.2f}".format)
purchase_analysis_gender_display
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 200]
age_ranges = ["<10", "10-14", "15-19", "20-24","25-29", "30-34", "35-39", "40+" ]
purchase_data_df["Age Ranges"] = pd.cut(purchase_data_df["Age"], bins, labels=age_ranges, include_lowest=True)
#purchase_data_df.head()
```

```python
age_data = purchase_data_df[["SN", "Age", "Age Ranges"]].drop_duplicates()
player_count_age = age_data["Age Ranges"].value_counts()
age_demo_df = pd.DataFrame({"Total Count": player_count_age}).sort_index()
age_demo_df["Percentage of Players"] = age_demo_df["Total Count"]/age_demo_df["Total Count"].sum()
age_demo_df_display = age_demo_df.copy()
age_demo_df_display["Percentage of Players"] = age_demo_df_display["Percentage of Players"].map("{:,.2%}".format)
age_demo_df_display["Total Count"] = age_demo_df_display["Total Count"].map("{:,}".format)
age_demo_df_display
```

```python
grouped_age = purchase_data_df.groupby("Age Ranges")
purchase_count_age = grouped_age["Purchase ID"].count()
total_revenue_age = grouped_age["Price"].sum()
avg_purchase_price_age = grouped_age["Price"].mean()
avg_purchase_per_person_age = (total_revenue_age / player_count_age)
purchase_analysis_age = pd.DataFrame({"Purchase Count": purchase_count_age,
                                        "Average Purchase Price": avg_purchase_price_age,
                                        "Total Purchase Value": total_revenue_age,
                                        "Avg Total Purchase per Person": avg_purchase_per_person_age,
                                        })
purchase_analysis_age_display = purchase_analysis_age.copy()
purchase_analysis_age_display["Average Purchase Price"] = purchase_analysis_age_display["Average Purchase Price"].map("${:,.2f}".format)
purchase_analysis_age_display["Total Purchase Value"] = purchase_analysis_age_display["Total Purchase Value"].map("${:,.2f}".format)
purchase_analysis_age_display["Avg Total Purchase per Person"] = purchase_analysis_age_display["Avg Total Purchase per Person"].map("${:,.2f}".format)
purchase_analysis_age_display
```

```python
grouped_spender = purchase_data_df.groupby("SN")
purchase_count_spender = grouped_spender["Purchase ID"].count()
total_revenue_spender = grouped_spender["Price"].sum()
avg_purchase_price_spender = grouped_spender["Price"].mean()
purchase_analysis_spender = pd.DataFrame({"Purchase Count": purchase_count_spender,
                                        "Average Purchase Price": avg_purchase_price_spender,
                                        "Total Purchase Value": total_revenue_spender,
                                        })
purchase_analysis_spender_display = purchase_analysis_spender.copy().sort_values(by=["Total Purchase Value"], ascending=False)
purchase_analysis_spender_display["Average Purchase Price"] = purchase_analysis_spender_display["Average Purchase Price"].map("${:,.2f}".format)
purchase_analysis_spender_display["Total Purchase Value"] = purchase_analysis_spender_display["Total Purchase Value"].map("${:,.2f}".format)
purchase_analysis_spender_display.head()
```

```python
grouped_item = purchase_data_df.groupby(["Item ID", "Item Name"])
purchase_count_item = grouped_item["Purchase ID"].count()
total_revenue_item = grouped_item["Price"].sum()
avg_purchase_price_item = grouped_item["Price"].mean()
purchase_analysis_item = pd.DataFrame({"Purchase Count": purchase_count_item,
                                        "Average Purchase Price": avg_purchase_price_item,
                                        "Total Purchase Value": total_revenue_item,
                                        })
purchase_analysis_item_display = purchase_analysis_item.copy().sort_values(by=["Purchase Count"], ascending=False)
purchase_analysis_item_display["Average Purchase Price"] = purchase_analysis_item_display["Average Purchase Price"].map("${:,.2f}".format)
purchase_analysis_item_display["Total Purchase Value"] = purchase_analysis_item_display["Total Purchase Value"].map("${:,.2f}".format)
purchase_analysis_item_display.head()
```
